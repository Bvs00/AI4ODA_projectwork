## Metaboloma

### Read Data Set

```python
import pandas as pd
import numpy as np

data_metaboloma_transpose = pd.read_excel("DataSet PAscale/Metaboloma.xlsx")
data_metaboloma_transpose

data_metaboloma = data_metaboloma_transpose.T

data_metaboloma.columns = data_metaboloma_transpose["Sample"]

data_metaboloma = data_metaboloma.iloc[1:]

data_metaboloma.columns = list(data_metaboloma.columns)

df_y_metaboloma = data_metaboloma["Label"].iloc[1:]

mz_metaboloma = data_metaboloma.iloc[0:1,1:]

df_y_qc_metaboloma = data_metaboloma["Label"].iloc[1:5]

df_x_metaboloma = data_metaboloma.iloc[1:,1:]

df_qc_metaboloma = data_metaboloma.iloc[1:5]

name_samples_metaboloma = df_x_metaboloma.transpose().columns

df_x_metaboloma
```

```python
# color mapping to use in plot
color_mapping = {
    'LC': 'red',
    'MC': 'green',
    'HC' : 'orange',
    'QC' : 'purple'
}

colors = [color_mapping[label] for label in df_y_metaboloma]
```

#### Split in Train e in Test

```python
from imblearn.over_sampling import BorderlineSMOTE

df_x_metaboloma_tic_noqc = df_x_metaboloma_tic.iloc[4:]
df_y_metaboloma_noqc = df_y_metaboloma.iloc[4:]

# bls = BorderlineSMOTE(random_state=42, k_neighbors=20)
# df_x_metaboloma_tic_over,df_y_metaboloma_over = bls.fit_resample(df_x_metaboloma_tic_noqc,df_y_metaboloma_noqc)

# # df_y_metaboloma_over.value_counts()
# print(df_y_metaboloma_over.value_counts())
print(df_y_metaboloma_noqc.value_counts())

plt.plot(df_y_metaboloma_noqc.value_counts())
plt.show()

x_tic = df_x_metaboloma_tic_noqc.copy()
y_tic = df_y_metaboloma_noqc.copy()

```

##### Division in test and train

```python
from sklearn.model_selection import train_test_split

x_train_tic, x_test_tic, y_train_tic, y_test_tic = train_test_split(x_tic, y_tic, test_size=0.3, stratify=df_y_metaboloma_noqc, random_state=99)
print(y_train_tic.value_counts())

name_sample_train_tic = x_train_tic.transpose().columns
```

#### Autoscaled
Autoscaled on the train data

```python
mean_tic= x_train_tic.mean(axis=0)
std_tic = x_train_tic.std(axis=0)
x_train_tic_centered = (x_train_tic.sub(mean_tic,axis=1)).div(std_tic,axis=1)

plt.scatter(mz_metaboloma, x_train_tic_centered.iloc[62:63])
plt.xlabel("M/Z")
plt.ylabel("Intensity")
plt.title("Samlple HC 11_M")
plt.show()
x_train_tic_centered[62:63].max(axis=1)
```

#### Elbow Graph

```python
from sklearn.decomposition import PCA

varianza_spiegata_tic = []
n_componenti_tic = np.arange(1, 53)

for n in n_componenti_tic:
    pca_elbow = PCA(n_components=n)
    pca_elbow.fit(x_train_tic_centered)
    varianza_spiegata_tic.append(np.sum(pca_elbow.explained_variance_ratio_))

plt.figure(figsize=(8, 6))
plt.plot(n_componenti_tic, varianza_spiegata_tic, marker='o')
plt.xlabel('# of principal components')
plt.ylabel('Explained Variance')
plt.title('Elbow Graph')
plt.grid(True)
plt.show()
```

#### PCA

```python
pca_tic = PCA(n_components=30)
x_train_tic_centered_pca = pca_tic.fit(x_train_tic_centered).transform(x_train_tic_centered)
x_train_tic_centered_pca
```

##### Try a Logistic Regression

```python
X_w_tic = pd.DataFrame(x_train_tic_centered_pca)
Y_w_tic = y_train_tic

def norm(x):
  return (x - train_stats['mean']) / train_stats['std']
X_w_tic.columns = [f'PC{i+1}' for i in range(X_w_tic.shape[1])]
X_w_tic = X_w_tic.transpose()
X_w_tic.columns = name_sample_train_tic
X_w_tic = X_w_tic.transpose()
X_w_tic
```

```python
from sklearn.model_selection import KFold,StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn import metrics as mt
from imblearn.over_sampling import BorderlineSMOTE

# x=df_x_lipidoma_tic_scaled
k = 45
# kf = KFold(k,shuffle=True,random_state=42)

skf = StratifiedKFold(n_splits=k)
bls = BorderlineSMOTE(random_state=99)
X_w_tic_temp,Y_w_tic_temp = bls.fit_resample(X_w_tic,Y_w_tic)

performance_scores = []
performance_scores_balanced = []
i = 0
# Itera attraverso i fold
for train_index, val_index in skf.split(X_w_tic_temp,Y_w_tic_temp):
    i+=1
    train_x_temp, test_x_temp = X_w_tic_temp.iloc[train_index],X_w_tic_temp.iloc[val_index]
    train_y_temp, test_y_temp = Y_w_tic_temp.iloc[train_index],Y_w_tic_temp.iloc[val_index]
    
    # print("FOLD %d" %i)
    #ribilanciamento
    # bls = BorderlineSMOTE(random_state=99)
    # train_x_temp,train_y_temp = bls.fit_resample(train_x_temp,train_y_temp)
    # print(train_y_temp.value_counts())

    # normalizzazione
    train_stats = pd.DataFrame()
    train_stats['mean'] = train_x_temp.mean(0)
    train_stats['std'] = train_x_temp.std(0)
    train_stats['std'][train_stats['std']==0] = 1

    #normalizzo i dati
    train_x_temp = norm(train_x_temp)
    test_x_temp = norm(test_x_temp)
    
    lm = LogisticRegression(penalty='l2', C=0.9, class_weight='balanced', multi_class='auto',max_iter=1000, n_jobs=4) # penalty{â€˜l1â€™, â€˜l2â€™, â€˜elasticnetâ€™, None}; multi_class{â€˜autoâ€™, â€˜ovrâ€™, â€˜multinomialâ€™}
    
    lm.fit(train_x_temp, train_y_temp)
    
    predicted_y_temp = lm.predict(test_x_temp)
    
    # Calcola e memorizza la misura delle prestazioni (es. accuratezza)
    accuracy_temp = np.round(mt.accuracy_score(test_y_temp,predicted_y_temp)*100,2)
    performance_scores.append(accuracy_temp)
    balanced_accuracy_temp = np.round(mt.balanced_accuracy_score(test_y_temp,predicted_y_temp)*100,2)
    performance_scores_balanced.append(balanced_accuracy_temp)
    
mean_accuracy_temp = sum(performance_scores)/k
mean_accuracy_balanced_temp = sum(performance_scores_balanced)/k

print("L'accuracy Ã¨: ", mean_accuracy_temp, "\nLa balanced_accuracy Ã¨: ", mean_accuracy_balanced_temp)
```
