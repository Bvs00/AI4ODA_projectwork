#### Define X and Y
Ancora non ho eliminato gli outlier.

Ancora non ho diviso train e test in modo che il train abbia i campioni con maggiore varianza.

```python
from imblearn.over_sampling import BorderlineSMOTE

df_x_metaboloma_tic_noqc = df_x_metaboloma_tic.iloc[4:]
df_y_metaboloma_noqc = df_y_metaboloma.iloc[4:]

print(df_y_metaboloma_noqc.value_counts())

x_tic = df_x_metaboloma_tic_noqc.copy()
y_tic = df_y_metaboloma_noqc.copy()

print("\nAfter drop the outleir in LC class\n")
outlier = x_tic.index[[34,67]]
x_tic = x_tic.drop(index=outlier)
y_tic = y_tic.drop(index=outlier)

name_x_tic = x_tic.transpose().columns
print(y_tic.value_counts())
plt.plot(df_y_metaboloma_noqc.value_counts())
plt.show()
```

#### Save the mean and standard deviation of the training set

```python
mean_tic= x_tic.mean(axis=0)
std_tic = x_tic.std(axis=0)

x_tic = (x_tic.sub(mean_tic,axis=1)).div(std_tic,axis=1)
```

#### Elbow Graph

```python
from sklearn.decomposition import PCA

varianza_spiegata_tic = []
n_componenti_tic = np.arange(1, 53)

for n in n_componenti_tic:
    pca_elbow = PCA(n_components=n)
    pca_elbow.fit(x_tic)
    varianza_spiegata_tic.append(np.sum(pca_elbow.explained_variance_ratio_))

plt.figure(figsize=(8, 6))
plt.plot(n_componenti_tic, varianza_spiegata_tic, marker='o')
plt.xlabel('# of principal components')
plt.ylabel('Explained Variance')
plt.title('Elbow Graph')
plt.grid(True)
plt.show()
```

#### PCA

```python
pca_tic = PCA(n_components=40)
pca_tic_fit= pca_tic.fit(x_tic)
x_tic_pca = pca_tic_fit.transform(x_tic)
```

#### Pre Models

```python
from imblearn.over_sampling import BorderlineSMOTE
X_w_tic = pd.DataFrame(x_tic_pca)
Y_w_tic = y_tic

X_w_tic.columns = [f'PC{i}' for i in range(X_w_tic.shape[1])]
X_w_tic = X_w_tic.transpose()
X_w_tic.columns = name_x_tic
X_w_tic = X_w_tic.transpose()

all_performance = []

#ribilanciamento
bls = BorderlineSMOTE(random_state=99)
x,y = bls.fit_resample(X_w_tic,Y_w_tic)
```

##### Fit the Logistic Regression model

```python
from sklearn.linear_model import LogisticRegression

FinalModel = LogisticRegression(penalty='l2', C=0.9, class_weight='balanced', multi_class='auto',max_iter=10000, n_jobs=4, solver='saga')
    
FinalModel.fit(x, y)

FinalModel
```

##### Fit the SVM model

```python
from sklearn.svm import SVC
from imblearn.over_sampling import BorderlineSMOTE

SVM_models =  SVC(C=0.9,kernel="linear")

SVM_models.fit(x,y)

SVM_models
```

##### Fit the RandomForest models

```python
from sklearn.ensemble import RandomForestClassifier

RF_model = RandomForestClassifier(n_estimators=200, random_state=3, max_depth=22, n_jobs=4)
RF_model.fit(x, y)

RF_model
```

## Analyze the model

#### Logistic Regression

```python
features_importance = pd.DataFrame({"PC": FinalModel.feature_names_in_, "Coefficient": abs(FinalModel.coef_[0])})
features_importance = features_importance.sort_values(by="Coefficient", ascending=False)

imp_features_pca = features_importance.loc[features_importance["Coefficient"]>=0.1]

name_pc = imp_features_pca.T.columns

features_imp_original = pd.DataFrame({"Feature": x_tic.columns,**{f"PC{i}":pca_tic_fit.components_[i]for i in name_pc}})
temp = features_imp_original.T
temp.columns = x_tic.columns
temp_2 = temp.T
features_imp_original = temp_2.iloc[:,1:]

most_important_features_lr = pd.DataFrame({"LogisticRegression":features_imp_original.mean(axis=1)}).sort_values(by="LogisticRegression",ascending=False)
most_important_features = pd.DataFrame({"LogisticRegression":features_imp_original.mean(axis=1)})
```

#### SVM

```python
features_importance = pd.DataFrame({"PC": SVM_models.feature_names_in_, "Coefficient": abs(SVM_models.coef_[0])})
features_importance = features_importance.sort_values(by="Coefficient", ascending=False)
features_importance
# choose the most important features
imp_features_pca = features_importance.loc[features_importance["Coefficient"]>=0.1]

name_pc = imp_features_pca.T.columns

features_imp_original = pd.DataFrame({"Feature": x_tic.columns,**{f"PC{i}":pca_tic_fit.components_[i]for i in name_pc}})
temp = features_imp_original.T
temp.columns = x_tic.columns
temp_2 = temp.T
features_imp_original = temp_2.iloc[:,1:]
most_important_features_svm = pd.DataFrame({"SVM":features_imp_original.mean(axis=1)}).sort_values(by="SVM",ascending=False)
most_important_features["SVM"] = features_imp_original.mean(axis=1)
most_important_features

most_important_features_svm
```

#### RandomForest

```python
features_importance = pd.DataFrame({"PC": RF_model.feature_names_in_, "Coefficient": abs(RF_model.feature_importances_)})
features_importance = features_importance.sort_values(by="Coefficient", ascending=False)
features_importance
# choose the most important features
imp_features_pca = features_importance.loc[features_importance["Coefficient"]>=0.1]

name_pc = imp_features_pca.T.columns

features_imp_original = pd.DataFrame({"Feature": x_tic.columns,**{f"PC{i}":pca_tic_fit.components_[i]for i in name_pc}})
temp = features_imp_original.T
temp.columns = x_tic.columns
temp_2 = temp.T
features_imp_original = temp_2.iloc[:,1:]

most_important_features_rf = pd.DataFrame({"RandomForest":features_imp_original.mean(axis=1)}).sort_values(by="RandomForest",ascending=False)
most_important_features["RandomForest"] = features_imp_original.mean(axis=1)
most_important_features_rf
```

```python
file_path = "most_important_features.xlsx"
most_important_features.to_excel(file_path)
```
