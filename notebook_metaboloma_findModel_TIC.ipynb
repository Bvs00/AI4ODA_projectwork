# Project Work
This project is composed from two dataset:
- Lipidoma
- Metaboloma

The aim of this project is identify bio-markers that allow to distinguish the evolution of pathology (HCV). We have 3 several class of patient:
- LC: Hepatocellular carcinoma
- HC: Hepatitis C virus
- MC: Mixed Cryoglobulinemia disease

#### Split in Train and in Test

##### Find the samples more distant from center

```python
lc_list = x_tic.iloc[0:67]      # LC
mc_list = x_tic.iloc[67:77]     # MC
hc_list = x_tic.iloc[77:100]    # HC
```

```python
center_lc = lc_.mean(axis=0)
center_mc = mc.mean(axis=0)
center_hc = hc.mean(axis=0)

distance_from_center_lc = pd.DataFrame({"Sample":lc_list.index,"Distance from center": np.linalg.norm(lc_-center_lc,axis=1)}).sort_values(by="Distance from center",ascending=False)
distance_from_center_mc = pd.DataFrame({"Sample":mc_list.index,"Distance from center": np.linalg.norm(mc-center_mc,axis=1)}).sort_values(by="Distance from center",ascending=False)
distance_from_center_hc = pd.DataFrame({"Sample":hc_list.index,"Distance from center": np.linalg.norm(hc-center_hc,axis=1)}).sort_values(by="Distance from center",ascending=False)
distance_from_center_lc
```

Split the data set in train and in test set

```python
train_size = 0.7
train_size_lc = int(len(distance_from_center_lc)*train_size); train_size_mc = int(len(distance_from_center_mc)*train_size); train_size_hc = int(len(distance_from_center_hc)*train_size)

list_train_lc = x_tic.loc[distance_from_center_lc["Sample"].iloc[0:train_size_lc]]
list_train_mc = x_tic.loc[distance_from_center_mc["Sample"].iloc[0:train_size_mc]]
list_train_hc = x_tic.loc[distance_from_center_hc["Sample"].iloc[0:train_size_hc]]

list_train_lc_y = y_tic.loc[distance_from_center_lc["Sample"].iloc[0:train_size_lc]]
list_train_mc_y = y_tic.loc[distance_from_center_mc["Sample"].iloc[0:train_size_mc]]
list_train_hc_y = y_tic.loc[distance_from_center_hc["Sample"].iloc[0:train_size_hc]]

x_train_tic = pd.concat([list_train_lc,list_train_mc,list_train_hc])
y_train_tic = pd.concat([list_train_lc_y,list_train_mc_y,list_train_hc_y])
y_train_tic
```

```python
x_test_tic = x_tic.loc[~x_tic.isin(x_train_tic.to_dict(orient='list')).all(axis=1)]
y_test_tic = y_tic[~y_tic.index.isin(y_train_tic.index)]
len(y_test_tic)
```

```python
name_sample_train_tic = x_train_tic.transpose().columns
name_sample_test_tic = x_test_tic.transpose().columns
```

#### Save the mean and standard deviation of the training set

```python
mean_tic= x_train_tic.mean(axis=0)
std_tic = x_train_tic.std(axis=0)
```

#### Elbow Graph

```python
from sklearn.decomposition import PCA

varianza_spiegata_tic = []
n_componenti_tic = np.arange(1, 53)

for n in n_componenti_tic:
    pca_elbow = PCA(n_components=n)
    pca_elbow.fit(x_train_tic)
    varianza_spiegata_tic.append(np.sum(pca_elbow.explained_variance_ratio_))

plt.figure(figsize=(8, 6))
plt.plot(n_componenti_tic, varianza_spiegata_tic, marker='o')
plt.xlabel('# of principal components')
plt.ylabel('Explained Variance')
plt.title('Elbow Graph')
plt.grid(True)
plt.show()
```

#### PCA

```python
pca_tic = PCA(n_components=10)
x_train_tic_pca = pca_tic.fit(x_train_tic).transform(x_train_tic)
x_train_tic_pca
```

#### Compose to Models

```python
X_w_tic = pd.DataFrame(x_train_tic_pca)
Y_w_tic = y_train_tic

X_w_tic.columns = [f'PC{i+1}' for i in range(X_w_tic.shape[1])]
X_w_tic = X_w_tic.transpose()
X_w_tic.columns = name_sample_train_tic
X_w_tic = X_w_tic.transpose()

all_performance = []
```

#### Try a Logistic Regression

```python
from sklearn.model_selection import LeaveOneOut
from sklearn.linear_model import LogisticRegression
from sklearn import metrics as mt
from imblearn.over_sampling import BorderlineSMOTE

loo = LeaveOneOut()

performance_scores = []
performance_scores_balanced = []
i = 0
# Itera attraverso i fold
for train_index, val_index in loo.split(X_w_tic,Y_w_tic):
    i+=1
    
    train_x_temp, test_x_temp = X_w_tic.iloc[train_index],X_w_tic.iloc[val_index]
    train_y_temp, test_y_temp = Y_w_tic.iloc[train_index],Y_w_tic.iloc[val_index]
    
    #ribilanciamento
    bls = BorderlineSMOTE(random_state=99)
    train_x_temp,train_y_temp = bls.fit_resample(train_x_temp,train_y_temp)

    # normalizzazione
    mean_tic_temp= train_x_temp.mean(axis=0)
    std_tic_temp = train_x_temp.std(axis=0)
    train_x_temp = (train_x_temp.sub(mean_tic_temp,axis=1)).div(std_tic_temp,axis=1)

    test_x_temp = (test_x_temp.sub(mean_tic_temp,axis=1)).div(std_tic_temp,axis=1)
    
    lm = LogisticRegression(penalty='l2', C=0.9, class_weight='balanced', multi_class='auto',max_iter=10000, n_jobs=4, solver='saga') # penalty{â€˜l1â€™, â€˜l2â€™, â€˜elasticnetâ€™, None}; multi_class{â€˜autoâ€™, â€˜ovrâ€™, â€˜multinomialâ€™}
    
    lm.fit(train_x_temp, train_y_temp)
    
    predicted_y_temp = lm.predict(test_x_temp)

    # Calcola e memorizza la misura delle prestazioni (es. accuratezza)
    accuracy_temp = np.round(mt.accuracy_score(test_y_temp.values,predicted_y_temp)*100,2)
    performance_scores.append(accuracy_temp)
    balanced_accuracy_temp = np.round(mt.balanced_accuracy_score(test_y_temp.values,predicted_y_temp)*100,2)
    performance_scores_balanced.append(balanced_accuracy_temp)
    
mean_accuracy_temp = sum(performance_scores)/i
mean_accuracy_balanced_temp = sum(performance_scores_balanced)/i

print("L'accuracy Ã¨: ", mean_accuracy_temp, "\nLa balanced_accuracy Ã¨: ", mean_accuracy_balanced_temp)

all_performance = [{"TIC-SUM":{
    "LogisticRegression_Validation": {
        "BorderlineSMOTE": mean_accuracy_balanced_temp
    }
}}]
```

##### Try on test set

```python
# autoscaling on test set
x_test_tic_centered = (x_test_tic.sub(mean_tic,axis=1)).div(std_tic,axis=1)
x_test_tic_centered_pca = pca_tic.transform(x_test_tic_centered)

x_test_tic_centered_pca = pd.DataFrame(x_test_tic_centered_pca)
x_test_tic_centered_pca.columns = [f'PC{i+1}' for i in range(x_test_tic_centered_pca.shape[1])]
x_test_tic_centered_pca = x_test_tic_centered_pca.transpose()
x_test_tic_centered_pca.columns = name_sample_test_tic
x_test_tic_centered_pca = x_test_tic_centered_pca.transpose()

predicted_y_test = lm.predict(x_test_tic_centered_pca)
balanced_accuracy = np.round(mt.balanced_accuracy_score(y_test_tic,predicted_y_test)*100,2)
all_performance[0]["TIC-SUM"]["LogisticRegression_Test"]= {"BorderlineSMOTE":balanced_accuracy}
balanced_accuracy
```

#### SVM

```python
from sklearn.model_selection import LeaveOneOut
from sklearn.svm import SVC
from sklearn import metrics as mt
from imblearn.over_sampling import BorderlineSMOTE

loo = LeaveOneOut()

performance_scores = []
performance_scores_balanced = []
i = 0
# Itera attraverso i fold
for train_index, val_index in loo.split(X_w_tic):
    i+=1
    train_x_temp, test_x_temp = X_w_tic.iloc[train_index],X_w_tic.iloc[val_index]
    train_y_temp, test_y_temp = Y_w_tic.iloc[train_index],Y_w_tic.iloc[val_index]
    
    # balancing
    train_x_temp,train_y_temp = BorderlineSMOTE(random_state=99).fit_resample(train_x_temp,train_y_temp)
    
    # normalization
    mean_tic_temp= train_x_temp.mean(axis=0)
    std_tic_temp = train_x_temp.std(axis=0)
    train_x_temp = (train_x_temp.sub(mean_tic_temp,axis=1)).div(std_tic_temp,axis=1)

    test_x_temp = (test_x_temp.sub(mean_tic_temp,axis=1)).div(std_tic_temp,axis=1)
    
    model_svc = SVC(C=0.9)
    model_svc.fit(train_x_temp,train_y_temp)
    
    predicted_y_temp = model_svc.predict(test_x_temp)
    
    accuracy_temp = np.round(mt.accuracy_score(test_y_temp,predicted_y_temp)*100,2)
    performance_scores.append(accuracy_temp)
    balanced_accuracy_temp = np.round(mt.balanced_accuracy_score(test_y_temp,predicted_y_temp)*100,2)
    performance_scores_balanced.append(balanced_accuracy_temp)
    
mean_accuracy_temp = sum(performance_scores)/i
mean_accuracy_balanced_temp = sum(performance_scores_balanced)/i

print("L'accuracy Ã¨: ", mean_accuracy_temp, "%\nLa balanced_accuracy Ã¨: ", mean_accuracy_balanced_temp,"%")

all_performance[0]["TIC-SUM"]["SVM_Validation"]= {"BorderlineSMOTE":mean_accuracy_balanced_temp}
```

##### Try on Test

```python
x_test_tic_centered = (x_test_tic.sub(mean_tic,axis=1)).div(std_tic,axis=1)
x_test_tic_centered_pca = pca_tic.transform(x_test_tic_centered)

x_test_tic_centered_pca = pd.DataFrame(x_test_tic_centered_pca)
x_test_tic_centered_pca.columns = [f'PC{i+1}' for i in range(x_test_tic_centered_pca.shape[1])]
x_test_tic_centered_pca = x_test_tic_centered_pca.transpose()
x_test_tic_centered_pca.columns = name_sample_test_tic
x_test_tic_centered_pca = x_test_tic_centered_pca.transpose()

predicted_y_test = model_svc.predict(x_test_tic_centered_pca)
balanced_accuracy = np.round(mt.balanced_accuracy_score(y_test_tic,predicted_y_test)*100,2)
all_performance[0]["TIC-SUM"]["SVM_Test"]= {"BorderlineSMOTE":balanced_accuracy}
balanced_accuracy
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics as mt
from sklearn.model_selection import LeaveOneOut
from imblearn.over_sampling import BorderlineSMOTE

loo = LeaveOneOut()

performance_scores = []
performance_scores_balanced = []
i = 0

for train_index, val_index in loo.split(X_w_tic):
    i+=1
    train_x_temp, test_x_temp = X_w_tic.iloc[train_index],X_w_tic.iloc[val_index]
    train_y_temp, test_y_temp = Y_w_tic.iloc[train_index],Y_w_tic.iloc[val_index]
    
    train_x_temp,train_y_temp = BorderlineSMOTE(random_state=99).fit_resample(train_x_temp,train_y_temp)
    
    # normalizzazione
    mean_tic_temp = train_x_temp.mean(axis=0)
    std_tic_temp = train_x_temp.std(axis=0)
    train_x_temp = (train_x_temp.sub(mean_tic_temp,axis=1)).div(std_tic_temp,axis=1)

    test_x_temp = (test_x_temp.sub(mean_tic_temp,axis=1)).div(std_tic_temp,axis=1)
    
    rf_model = RandomForestClassifier(n_estimators=200, random_state=3, max_depth=22, n_jobs=4)
    rf_model.fit(train_x_temp, train_y_temp)
    predicted_y_temp = rf_model.predict(test_x_temp)
    
    accuracy_temp = np.round(mt.accuracy_score(test_y_temp,predicted_y_temp)*100,2)
    performance_scores.append(accuracy_temp)
    balanced_accuracy_temp = np.round(mt.balanced_accuracy_score(test_y_temp,predicted_y_temp)*100,2)
    performance_scores_balanced.append(balanced_accuracy_temp)
    
mean_accuracy_temp = sum(performance_scores)/i
mean_accuracy_balanced_temp = sum(performance_scores_balanced)/i

print("L'accuracy Ã¨: ", mean_accuracy_temp, "%\nLa balanced_accuracy Ã¨: ", mean_accuracy_balanced_temp,"%")

all_performance[0]["TIC-SUM"]["RandomForests_Validation"]= {"BorderlineSMOTE":mean_accuracy_balanced_temp}
```

##### Try on Test

```python
x_test_tic_centered = (x_test_tic.sub(mean_tic,axis=1)).div(std_tic,axis=1)
x_test_tic_centered_pca = pca_tic.transform(x_test_tic_centered)

x_test_tic_centered_pca = pd.DataFrame(x_test_tic_centered_pca)
x_test_tic_centered_pca.columns = [f'PC{i+1}' for i in range(x_test_tic_centered_pca.shape[1])]
x_test_tic_centered_pca = x_test_tic_centered_pca.transpose()
x_test_tic_centered_pca.columns = name_sample_test_tic
x_test_tic_centered_pca = x_test_tic_centered_pca.transpose()

predicted_y_test = rf_model.predict(x_test_tic_centered_pca)
balanced_accuracy = np.round(mt.balanced_accuracy_score(y_test_tic,predicted_y_test)*100,2)
all_performance[0]["TIC-SUM"]["RandomForests_Test"]= {"BorderlineSMOTE":balanced_accuracy}
balanced_accuracy
```
